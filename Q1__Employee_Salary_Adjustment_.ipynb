```python
import numpy as np
import pandas as pd

# Initial salary data
salaries = np.array([45000, 52000, 61000, 48000, 75000])
print("Original Salaries:", salaries)

# 1. Increase all salaries by 10%
adjusted_salaries = salaries + (salaries * 0.10)
print("Adjusted Salaries (10% increase):", adjusted_salaries)

# 2. Find the average salary

average_salary = np.mean(adjusted_salaries)
print("Average Salary: ",average_salary)

# 3. Find the maximum and minimum salary

max_salary = np.max(adjusted_salaries)
min_salary = np.min(adjusted_salaries)
print("Maximum Salary:", max_salary)
print("Minimum Salary: ", min_salary)

# 4. Print salaries greater than 60,000

high_salaries = adjusted_salaries[adjusted_salaries > 60000]
print("Salaries greater than 60,000 ", high_salaries)
```
