```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv("/content/drive/MyDrive/ICT-IIT Palakad- Advanced AI/data/StudentsPerformance.csv")
```

```python
#1. Calculate the mean, median, and mode for the math, reading, and writing scores
for col in ['math score', 'reading score', 'writing score']:
    print(f"\n{col.upper()}")
    print("Mean:", df[col].mean())
    print("Median:", df[col].median())
    print("Mode:", df[col].mode()[0])
```

```python
#2. Determine the range, variance, and standard deviation for each score category
for col in ['math score', 'reading score', 'writing score']:
    print(f"\n{col.upper()}")
    print("Range:", df[col].max() - df[col].min())
    print("Variance:", df[col].var())
    print("Std Dev:", df[col].std())
```

```python
#3. Identify the mean, median, and mode of math scores for students with different parental education levels
grouped = df.groupby('parental level of education')['math score']
#print(df.groupby('parental level of education')['math score'])
print("\nMean:\n", grouped.mean())
print("\nMedian:\n", grouped.median())
print("\nMode:\n", grouped.apply(lambda x: x.mode()[0]))
```

```python
#4. 4. Compare the mean scores of students who took the test preparation course versus those who didn’t.
print(df.groupby('test preparation course')['math score'].mean())
```

```python
#5. Calculate the interquartile range (IQR) for each score category
for col in ['math score', 'reading score', 'writing score']:
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  print(f"{col} IQR:", Q3 - Q1)
```

```python
#6. Determine the standard deviation for math scores by lunch type (free/reduced vs.standard).
print(df.groupby('lunch')['math score'].std())
```

```python
#7.Find the correlation coefficient between math and reading scores, and then between reading and writing scores.
corr_math_read = df['math score'].corr(df['reading score'])
corr_read_write = df['reading score'].corr(df['writing score'])

print("Math vs Reading:", corr_math_read)
print("Reading vs Writing:", corr_read_write)
```

```python
#9.Calculate the covariance between math and reading scores. How does it compare with the correlation? What does it suggest?
cov1 = np.cov(df['math score'], df['reading score'])[0][1]
print("Covariance (Math vs Reading):", cov1)
cov2 = np.cov(df['writing score'], df['reading score'])[0][1]
print("Covariance (writing vs Reading):", cov2)
```

```python
#10.  Z test
population_mean = 50

sample_mean = df['math score'].mean()
sample_std = df['math score'].std()
n = len(df)

z = (sample_mean - population_mean) / (sample_std / np.sqrt(n))

print("Z-score:", z)
```

```python
#11. T -test
standard = df[df['lunch'] == 'standard']['reading score']
free = df[df['lunch'] == 'free/reduced']['reading score']

t_stat, p_val = stats.ttest_ind(standard, free)

print("T-stat:", t_stat)
print("p-value:", p_val)
```

```python
#12. F -test
completed = df[df['test preparation course'] == 'completed']['math score']
none = df[df['test preparation course'] == 'none']['math score']

f_stat = np.var(completed, ddof=1) / np.var(none, ddof=1)

print("F-statistic:", f_stat)
```
