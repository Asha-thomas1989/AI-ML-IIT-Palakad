```python
import numpy as np
import pandas as pd

data = {
'Employee':['A','B','C','D','E','F'],
'Department':['HR','IT','IT','Finance','HR','Finance'],
'Salary':[50000,65000,70000,72000,52000,68000],
'Experience':[2,5,7,6,3,4]
}

#1. data frame
df= pd.DataFrame(data)
print("\n---Data Frame ---")
print(df)

# 2. basic statistics
print("\n\n---Basic Sattistics---")
print(df.describe())

#3. Average salary per department

print("\n\n Total Salary Department wise :")
df_sum =df.groupby('Department') ['Salary'].sum()
print(df_sum)

print("\n\n Mean Salary Department wise :")
df_mean =df.groupby('Department') ['Salary'].mean()
print(df_mean)
# 4. Employees with exp greater than 4 years
df_exp = df[df['Experience']> 4]
print("\n\n Employess with experience greater than 4 years \n ")
print(df_exp['Employee'])

# 5. Add a new coloumn Bonous =10% of salary

df['Bonous']= df['Salary']*0.10
print("\n--- Updated DataFrame with Bonus ---")
print(df)
```
