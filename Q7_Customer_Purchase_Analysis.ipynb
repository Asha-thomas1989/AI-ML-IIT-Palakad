```python
import numpy as np
import pandas as pd
```

```python

data = {
'CustomerID':[101,102,103,104,105,106,107],
'Age':[25,45,35,50,23,40,60],
'PurchaseAmount':[250,600,400,800,150,700,900],
'City':['Delhi','Mumbai','Delhi','Chennai','Mumbai','Delhi','Chennai']
}
```

```python
df_data = pd.DataFrame(data)
df_data
```

```python
# 1. Find the average purchase amount by city.
df_data.groupby('City')['PurchaseAmount'].mean()
```

```python
# 2. Identify customers with purchase above overall average.
df_average =df_data['PurchaseAmount'].mean()
print(" Average purchase : ", df_average)
df_highcustomer = df_data[df_data['PurchaseAmount']> df_average]
print("Customers are :\n ")
print(df_highcustomer)
```

```python
#3 Find the city generating the highest revenue.
df_city = df_data.groupby('City')['PurchaseAmount'].sum()
print("Total purchase of each city:\n \n  ", df_city)
df_city_idx = df_city.idxmax()
print("\n")
print("Highest Revenue City is ", df_city_idx)
print("Purchase amount is ", df_city[df_city_idx])
```

```python
# 4. Create a column Age_Group:

for i in range(len(df_data)):
  if (df_data['Age'][i]< 30):
    df_data.at[i,'Age Group'] ='young'
  elif (df_data['Age'][i] >= 30 and df_data['Age'][i] <= 50):
    df_data.at[i,'Age Group'] ='Adult'
  elif (df_data['Age'][i] > 30 and df_data['Age'][i] >50):
    df_data.at[i,'Age Group'] ='Senior'

df_data
```

```python
#5. Find average purchase per age group.
df_data.groupby('Age Group')['PurchaseAmount'].mean()
```
