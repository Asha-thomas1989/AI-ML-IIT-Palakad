```python
import numpy as np
import pandas as pd
```

```python
dates = pd.date_range('2024-01-01', periods=12, freq='M')
sales = [200,220,250,270,300,320,350,360,380,400,420,450]
df = pd.DataFrame({'Date':dates,'Sales':sales})
df
```

```python
# 1. Convert Date to datetime index
# Setting the index allows for time-based slicing and plotting
#df.set_index('Date', inplace=True)

# 2. Compute monthly percentage growth
# (Current - Previous) / Previous * 100
df['Monthly_Growth (%)'] = df['Sales'].pct_change() * 100


# 3. Calculate 3-month rolling average sales
# 'window=3' looks at the current month and the two previous months
df['3_Month_Rolling_Avg'] = df['Sales'].rolling(window=3).mean()

# 4. Identify the month with highest growth rate
# We skip the first month (NaN) since it has no previous month to compare to
highest_growth_month = df['Monthly_Growth (%)'].idxmax()
highest_growth_value = df['Monthly_Growth (%)'].max()

print("--- Time Series Analysis Table ---")
print(df)
print(f"\nMonth with highest growth: {highest_growth_month.strftime('%B %Y')}")
print(f"Growth Rate: {highest_growth_value:.2f}%")
```
