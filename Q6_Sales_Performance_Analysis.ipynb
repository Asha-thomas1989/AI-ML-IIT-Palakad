```python
import numpy as np
import pandas as pd
```

```python
sales_data = {
'Region':['North','South','East','West','North','South','East','West'],
'Salesperson':['A','B','C','D','E','F','G','H'],
'Sales':[12000,15000,13000,17000,16000,14000,15500,16500]
}
```

```python
df_data = pd.DataFrame(sales_data)
print('--- Initial data Frame ----\n ')
print(df_data)
```

```python
# 1. Total Sales per region

print(" Sales per Region")
df_data.groupby('Region')['Sales'] .sum()
```

```python
#2. Top Sales Person
print(" Top Sales person:")
top_idx = df_data['Sales'].idxmax()
#print(top_idx)
print(df_data['Salesperson'][top_idx])
```

```python
# 3. Average Sales across Regions
print("Average sales across regions \n ")
df_data.groupby('Region')['Sales'].mean()
```

```python
# 4. Sort the data set by sales ascending
print(" Sales in Ascending order")
df_data.sort_values(by='Sales', ascending=True)
```
