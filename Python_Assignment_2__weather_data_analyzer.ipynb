```python
print("--- Enter Weather Data ---")
region = input("Enter the Region name: ")
temperature = float(input(f"Enter the current temperature in {region} (°C): "))
humidity = float(input(f"Enter the humidity percentage in {region} (%): "))

# d. Function to print the stored results
def print_summary(reg, temp, hum):
    print(f"\n--- Analysis for {reg} ---")
    print(f"Temperature: {temp}°C")
    print(f"Humidity: {hum}%")

# c. Function with conditional statements to determine weather conditions
def determine_condition(temp, hum):
    if temp >= 30 and hum <= 40:
        return "Sunny and Dry"
    elif hum > 70:
        return "Rainy or Very Humid"
    elif temp < 15:
        return "Cold"
    else:
        return "Moderate/Clear"

# --- Main Execution ---
# Call the functions to process the user's data
print_summary(region, temperature, humidity)

# Store and print the condition
current_condition = determine_condition(temperature, humidity)
print(f"Weather Condition: {current_condition}")
```
