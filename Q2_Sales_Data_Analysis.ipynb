```python
import numpy as np
# sales data for 10 days
sales = np.array([120,150,180,200,170,160,210,220,190,175])
#1. find total sales
sales_total = np.sum(sales)
print("Total Sales",sales_total)

#2. Mean and standard deviation
sales_mean = np.mean(sales)
sales_std =sales.std()
print(" Mean Sales",sales_mean)
print(" Satandard deviation",sales_std)

#3. days where Sales exceeded the average
for i in range (len(sales)):
  if (sales[i] > sales_mean):
    print("Day " , i+1, " sales was",sales[i])

# 4. normalize using min or max scaling
sales_min =sales.min()
sales_max = sales .max()
print("min sales ", sales_min)
print("max sales ", sales_max)

normalized_sales = (sales - sales_min) / (sales_max - sales_min)
print("normalized sales", normalized_sales)
```
