```python
account_holder = input("Enter Account Holder Name: ")
balance = 0.0  # Starting balance (Float)

# d. Functions for bank operations
def display_balance():
    print(f"\n[Balance Enquiry] Account: {account_holder} | Current Balance: ${balance:.2f}")

def deposit():
    global balance
    amount = float(input("Enter amount to deposit: "))
    if amount > 0:
        balance += amount
        print(f"Success: ${amount:.2f} deposited.")
    else:
        print("Error: Amount must be positive.")

def withdraw():
    global balance
    amount = float(input("Enter amount to withdraw: "))
    # c. Conditional statement for sufficient balance check
    if amount > balance:
        print(f"Transaction Denied: Insufficient funds! (Available: ${balance:.2f})")
    elif amount <= 0:
        print("Error: Amount must be positive.")
    else:
        balance -= amount
        print(f"Success: ${amount:.2f} withdrawn.")

# --- a. Simulation Loop ---
while True:
    print("\n--- BANK MENU ---")
    print("1. Deposit Money")
    print("2. Withdraw Money")
    print("3. Check Balance")
    print("4. Exit")

    choice = input("Select an option (1-4): ")

    if choice == '1':
        deposit()
    elif choice == '2':
        withdraw()
    elif choice == '3':
        display_balance()
    elif choice == '4':
        print(f"Thank you for banking with us, {account_holder}. Goodbye!")
        break  # Exits the loop and terminates the program
    else:
        print("Invalid choice! Please select 1, 2, 3, or 4.")
```
