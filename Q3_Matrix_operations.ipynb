```python
# Sales for 3 products across 4 regions
import numpy as np
sales_matrix = np.array([
[120,140,130,150], #product 1
[80,90,100,95], # product 2
[200,210,190,205]]) # product 3
```

```python
# 1. total sales per product
Total_sales_product = sales_matrix.sum(1)
print("total sales product wise", Total_sales_product)

# 2. total sales per region
Total_sales_region = sales_matrix.sum(0)
print("total sales region wise", Total_sales_region)

#3. best performing product

best_product_idx = np.argmax(Total_sales_product)
print("Best performing product: ",(best_product_idx + 1))

# 4. Identify the region with highest sales
best_region_idx = np.argmax(Total_sales_region)
print("Region with highest sales: ", (best_region_idx + 1))
```
