```python
def get_weather_condition(temp, humidity):
    """c. Implement conditional statements to determine weather conditions"""
    if temp > 30 and humidity < 40:
        return "Sunny and Dry"
    elif humidity > 70:
        return "Rainy/Humid"
    elif temp < 15:
        return "Cold"
    else:
        return "Moderate/Clear"

def calculate_averages(temp_list, hum_list):
    """d. Use functions to calculate average temperature and humidity"""
    avg_temp = sum(temp_list) / len(temp_list)
    avg_hum = sum(hum_list) / len(hum_list)
    return avg_temp, avg_hum

# --- Main Program ---
# a. Analyze weather data for a given region
region = input("Enter the name of the Region: ")

# b. Use variables and data types (Lists) to store 7 days of data
temperatures = []
humidities = []

print(f"\n--- Entering 7-day data for {region} ---")

for i in range(1, 8):
    t = float(input(f"Day {i} - Enter Temperature (°C): "))
    h = float(input(f"Day {i} - Enter Humidity (%): "))

    temperatures.append(t)
    humidities.append(h)

    # c. Determine and print condition for each specific day
    condition = get_weather_condition(t, h)
    print(f" > Day {i} Analysis: {condition}")

# d. Perform final calculations and print results
average_t, average_h = calculate_averages(temperatures, humidities)

print(f"\n--- FINAL WEATHER REPORT: {region.upper()} ---")
print(f"Average Weekly Temperature: {average_t:.2f}°C")
print(f"Average Weekly Humidity: {average_h:.2f}%")

# Extra: Determine overall weekly climate
overall_condition = get_weather_condition(average_t, average_h)
print(f"Overall Weekly Climate: {overall_condition}")
```
