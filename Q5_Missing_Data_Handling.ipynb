```python
import numpy as np
import pandas as pd
```

```python
data = {
'Name':['John','Anna','Peter','Linda','David'],
'Age':[28, np.nan, 35, 32, np.nan],
'Salary':[50000,54000,np.nan,58000,60000]
}
```

```python
df_data= pd.DataFrame(data)
print("\n  ---Initial Data Frame  ---")
print(df_data)
```

```python
#1. Detect Missing Values
print("Missing Values Display \n ")
print(df_data.isna())
print("no of values missing in each coloumn \n ")
print(df_data.isna().sum())
```

```python
# 2. Replace misisng age with mean age
df_data['Age'].fillna(df_data['Age'].mean() ,inplace = True)
print(" Data after replacing missing age with mean value")
print(df_data)
```

```python
# 3. Replace missing Salary with median salary
df_data['Salary'].fillna(df_data['Salary'].median() , inplace =True)
print("\n --Data after replacing missing salary with Median--")
print(df_data)
```

```python
# 4.Remove rows with any remianing mising values
df_clean = df_data.dropna()
print(" \n Data after removal of any remianing missing values \n ")
print(df_clean)
```
